```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
x=np.array([1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6])
y=np.array([0.169610272,0.283395813,0.386358738,0.470227872,0.433281294,0.600267648,0.73833898,0.79031502,0.877464268,0.843564462,0.964438917])
```

```python
X = x.reshape(-1, 1)
modelo=LinearRegression()
modelo.fit(X,y)
```

```python
y_pred=modelo.predict(X)
r_squared=r2_score(y,y_pred)
print("coeficiente de determinación:",r_squared)
```

```python
df = pd.DataFrame({'Voltaje': x, 'Corriente': y})
sns.lmplot(x='Voltaje', y='Corriente', data=df)
```
