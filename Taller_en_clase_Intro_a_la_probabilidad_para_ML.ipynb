## Primer punto Distribuciones discretas.
1. Cree un arreglo de N muestras con N a su selección, pero mayor a 1000, de números distribuidos con las correspondientes distribuciones, para cada arreglo de números:
1. Grafique los números aleatorios,
2. Ordene de menor a mayor y grafique,
4. Encuentre los valores teoricos de media y varianza, y comparelos con los estimados.

```python
#Para todas las simulaciones:
N=1000
m=30
numpy.random.seed(seed=2**32 - 1)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```python
# bernoulli con cierto p entre 0 y 1
p=0.3
from scipy.stats import bernoulli
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
r = bernoulli.rvs(p, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
# Grafico Experimentos
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
r.sort()
# Grafico Experimentos ordenados
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli ordenados con p="+str(p))
plt.show()
# Histograma
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
# Datos estadisticos
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la bernoulli con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la bernoulli con p=",p,"son u_est=" ,avg,"var_est=",varest)


```


> La distribución Bernoulli permite analizar la probabilidad de éxito o fracaso de ejecutar un experimento, para esta ocasión ambos casos tienen igual probabilidad (0.5), la simulación se encarga de hacer 1000 experimentos que pueden arrojar un valor de 0 o 1 como únicas opciones de resultado.

> Este tipo de distribución se pueden aplicar en distintas situaciones como por ejemplo en el caso clásico de arrojar una moneda, donde puede salir solo cara o sello. Otro caso del modelo Bernoulli es en el área de las ciencias de la salud el resultado de suministrar un medicamento, funciona o no funciona.

```python
# binomial con cierto p entre 0 y 1 y cierto n
n,p = 10, 0.5
from scipy.stats import binom
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
r = binom.rvs(n,p, size=N)
# Grafico Experimentos
plt.plot(r)
plt.title(str(N)+" Experimentos distribución Binomial con p="+str(p)+" y " +str(n) +" ensayos")
plt.show()
r.sort()
# Grafico Experimentos ordenados
plt.plot(r)
plt.title(str(N)+" Experimentos ordenados distribución Binomial con p="+str(p)+" y " +str(n) +" ensayos")
plt.show()
# Histograma
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos Binomial con p="+str(p)+" y " +str(n) +" ensayos")
plt.show()
# Datos estadisticos
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la binomial con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la binomial con p=",p,"son u_est=" ,avg,"var_est=",varest)
```


El modelo probabilístico binomial de forma similar al Bernoulli dado que toma la probabilidad de éxito o fracaso de un evento, con la diferencia en que plantea es la idea de cuantos éxitos pueden ocurrir en n ensayos. Por ello es por lo que la simulación hace mil intentos con 10 ensayos cada uno y generando así un valor de cuantos éxitos ocurren en cada caso, se logra evidenciar que, en los valores lejanos a la media, la ocurrencia es poca o nula en algunas simulaciones.

Un ejemplo de su uso en la ingeniería se refleja en la producción de un producto y se busca saber la probabilidad de que entre determinada cantidad de productos se pueda hallar alguno defectuoso.


```python
# Geométrica con cierto p entre 0 y 1
p = 0.3
from scipy.stats import geom
mean, var, skew, kurt = geom.stats(p, moments='mvsk')
r = geom.rvs(p, size=N)
# Grafico Experimentos
plt.plot(r)
plt.title(str(N)+" puntos distribución Geometrica con p="+str(p))
plt.show()
r.sort()
# Grafico Experimentos ordenados
plt.plot(r)
plt.title(str(N)+" puntos distribución Geometrica ordenados con p="+str(p))
plt.show()
# Histograma
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos geometrica con p="+str(p))
plt.show()
# Datos estadisticos
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la Geometrica con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la geometrica con p=",p,"son u_est=" ,avg,"var_est=",varest)
```


La distribución geométrica consiste en tener que repetir n cantidad de veces algún experimento, entonces lo que acontece es que realiza el experimento de Bernoulli tantas veces sea necesario hasta obtener el resultado que se requiere. De las simulaciones se logra evidenciar que el éxito se consigue normalmente en pocos intentos, las situaciones en las que toma mas de ocho ensayos no son tan comunes. Contrastando de forma correcta con los datos estadísticos, donde se establece que la media se halla alrededor de tres. También se destaca que los datos no se desvían de forma simétrica, por debajo de la media los datos se acumulan en pocos valores, mientras que por encima se distribuyen en más experimentos.

Un ejemplo práctico de esta distribución se halla en los sistemas de búsqueda en línea donde se analiza la cantidad de clics requeridos hasta que se llega a la respuesta que requiere el usuario.


```python
# Poisson con cierto Lamnda L
L = 5
from scipy.stats import poisson
mean, var, skew, kurt =poisson.stats(L, moments='mvsk')
r = poisson.rvs(L, size=N)
# Grafico Experimentos
plt.plot(r)
plt.title(str(N)+" puntos distribución Geometrica con Lamnda="+str(L))
plt.show()
r.sort()
# Grafico Experimentos ordenados
plt.plot(r)
plt.title(str(N)+" puntos distribución Poisson ordenados con Lamnda="+str(L))
plt.show()
# Histograma
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos Poisson con Lamnda="+str(L))
plt.show()
# Datos estadisticos
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la Poisson con Lamnda=",L,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la Poisson con Lamnda=",L,"son u_est=" ,avg,"var_est=",varest)
```


El modelo Poisson lo que establece es que a partir del parámetro lambda simula la ocurrencia de eventos en un tiempo determinado. En el entorno simulado se ha de apreciar que dicha distribución empieza a tener un comportamiento gaussiano mas notorio, de tal manera que la ocurrencia de sus eventos se concentra hacia la media y aparecen dos colas donde la ocurrencia es mínima.

Para el caso de la ingeniería, esta distribución sirve para estimar la calidad de un producto a partir de cuantos pueden mostrarse defectuosos en un intervalo de tiempo.


```python
#Binomial negativa con parametro p después de r fallos
R,p = 5,0.5
from scipy.stats import nbinom
mean, var, skew, kurt = nbinom.stats(R, p, moments='mvsk')
r = nbinom.rvs(R,p, size=N)
# Grafico Experimentos
plt.plot(r)
plt.title(str(N)+" Experimentos distribución Binomial negativa con p="+str(p)+" y " +str(R) +" exitos")
plt.show()
r.sort()
# Grafico Experimentos ordenados
plt.plot(r)
plt.title(str(N)+" Experimentos ordenados distribución Binomialnegativas con p="+str(p)+" y " +str(R) +" exitos")
plt.show()
# Histograma
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos Binomial negativa con p="+str(p)+" y " +str(R) +" exitos")
plt.show()
# Datos estadisticos
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la binomial negativa con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la binomial negativa con p=",p,"son u_est=" ,avg,"var_est=",varest)
```


Por su parte la binomial negativa se encarga analizar la cantidad de repetir el experimento las veces necesarias hasta tener los éxitos requeridos, como es natural en las distribuciones discretas se basan en el modelo de Bernoulli para realizar su análisis de éxito y error. La grafica también muestra una gran concentración de valores entorno a la media y un comportamiento gaussiano, solo que no se halla tan simétrico teniendo una cola mucho mas larga.

En un caso de ingeniería se realiza la estimación de fallos necesarios de componentes para que falle todo un sistema. En un caso biológico, se puede estudiar la tasa de reproducción de una especie, la cantidad de intentos que le lleva a una especie reproducirse para generar la descendencia que sustente la siguiente generación de la especie.


## Segundo punto Distribuciones contínuas.


```python
#Para todas las simulaciones:
N=1000
m=30
numpy.random.seed(seed=0)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```python
# Normal de media 5 y desviación estandar 0.5
u=5
std=0.5
mean, var, skew, kurt = scipy.stats.norm.stats(moments='mvsk',loc=u,scale=std)
r = scipy.stats.norm.rvs (loc=u,scale=std, size=N)
# Grafico Experimentos
plt.plot(r)
plt.title(str(N)+" puntos distribuidos normales con media="+str(u)+" y std="+str(std))
plt.show()
r.sort()
# Grafico Experimentos ordenados
plt.plot(r)
plt.title(str(N)+" puntos distribuidos normales ordenados con u="+str(u)+ " y std= "+str(std))
plt.show()
# Histograma
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos normales con u="+str(u)+ " y std= "+str(std))
plt.show()
# Datos estadisticos
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la normal son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```python
# Uniforme entre 0 y 255
from scipy.stats import uniform
mean, var, skew, kurt = uniform.stats(moments='mvsk')
r = uniform.rvs (loc=1,scale=255, size=N)
# Grafico Experimentos
plt.plot(r)
plt.title(str(N)+" puntos distribuidos Uniforme entre 0 y 255")
plt.show()
r.sort()
# Grafico Experimentos ordenados
plt.plot(r)
plt.title(str(N)+" puntos distribuidos Uniforme entre 0 y 255")
plt.show()
# Histograma
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos Uniforme entre 0 y 255")
plt.show()
# Datos estadisticos
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la uniforme son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```


La distribución uniforme, por definición se conoce que todo su rango de elementos tiene la misma probabilidad de ocurrencia. La simulación permite ver que en el rango hay una ocurrencia de salida elevada de todos los valores, de esa manera al ver la gráfica de distribución ordenada tiende a ser una línea recta, de forma similar a la teoría de la función acumulada.

En la ingeniería electrónica se puede aprovechar dicha distribución para tener un sistema de generación de números aleatorios en un rango, todos equiprobables y con ello probar fluctuaciones en señales de comunicación o en ensayos de los componentes en distintas circunstancias.


```python
#T de Student con paramero K
K=1.0
from scipy.stats import t
mean, var, skew, kurt = t.stats(K,moments='mvsk')
r = t.rvs (K,loc=0,scale=1, size=N)
# Grafico Experimentos
plt.plot(r)
plt.title(str(N)+" puntos distribuidos T de Student con parametro K="+str(K))
plt.show()
r.sort()
# Grafico Experimentos ordenados
plt.plot(r)
plt.title(str(N)+" puntos distribuidos T de Student con parametro K="+str(K))
plt.show()
# Histograma
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos T de Student con parametro K="+str(K))
plt.show()
# Datos estadisticos
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la T student son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

La distribución t student se comporta de manera similar a una distribución normal con la diferencia de que su varianza es desconocida al ser un estudio de una muestra de la población, donde dicha muestra es muy pequeña en comparación. Se sabe que los grados de libertad representa el tamaño en comparación al de la población al ser en este caso 1 hace que la distribución se asemeje a una cauchy, cullas colas son infinitas, dando así que sus valores extremos sean mas probables de ocurrir, tal y como se ve en la distribución ordenada.

Las aplicaciones de la t studen sirve para estudiar intervalos de confianza en las situaciones donde no se conoce la varianza poblacional, por ejemplo: se tiene un cargamento de resistencias y se desea conocer si se cumplen con las especificaciones, para ello se toma una muestra y se mide, la desviación estándar con el total de resistencias es desconocido.


```python
# Exponencial con Lamnda L
L=0.3
from scipy.stats import expon
mean, var, skew, kurt = expon.stats(L,moments='mvsk')
r = expon.rvs (L,scale=1, size=N)
# Grafico Experimentos
plt.plot(r)
plt.title(str(N)+" puntos distribuidos exponencial con Lamnda="+str(L))
plt.show()
r.sort()
# Grafico Experimentos ordenados
plt.plot(r)
plt.title(str(N)+" puntos distribuidos exponencial con Lamnda="+str(L))
plt.show()
# Histograma
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos exponencial con Lamnda="+str(L))
plt.show()
# Datos estadisticos
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la exponencial son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

la exponencial sirve para modelar el crecimiento o decrecimiento de una cantidad respecto al tiempo o intervalo. Para este caso al tener una lambda positiva sugiere un crecimiento lento en intervalos, lo que se traduce en que los eventos son menos probables al ir avanzando en la variable independiente, como se ve en el histograma los valores bajos tienen mayor ocurrencia que los altos.

Una utilidad de dicha distribución es en la biología el estudio de la mortalidad de la población, si se halla una tasa de mortandad se puede estudiar que tan probable es que un ser pase de ese tiempo.


```python
#Chi cuadrada con K grados de libertad
K=3.0
from scipy.stats import chi2
mean, var, skew, kurt = chi2.stats(K,moments='mvsk')
r = chi2.rvs (K,loc=0,scale=1, size=N)
# Grafico Experimentos
plt.plot(r)
plt.title(str(N)+" puntos distribuidos chi cuadrada con "+str(K)+" grados de libertad")
plt.show()
r.sort()
# Grafico Experimentos ordenados
plt.plot(r)
plt.title(str(N)+" puntos distribuidos  chi cuadrada con "+str(K)+" grados de libertad")
plt.show()
# Histograma
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos  chi cuadrada con "+str(K)+" grados de libertad")
plt.show()
# Datos estadisticos
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la chi cuadrada son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

La distribución chi cuadrada simulada tiene pocos grados de libertad lo que hace que no se vea casi la forma de la campana, pero se cumplen las características, al ser una distribución con cuadrados no hay valores del lado negativo, ni en cero. El bajo grado de libertad hace que la cola derecha de la distribución sea amplia dando así paso a mayor probabilidad de ocurrencia a los valores extremos. Por lo cual, la grafica ordenada muestra una notoria creciente al final.

En ingeniería un ejemplo de esta distribución viene siendo al comprar un lote de resistencias ponerlas a prueba antes de integrarlas a un gran circuito, probando si funcionan en el rango de frecuencia requerido, entonces se usa la chi cuadrado para verificar con un valor crítico.


```python
#Gamma con parametros k y theta t
k,t=2.54,0.43
from scipy.stats import gamma
mean, var, skew, kurt = gamma.stats(K,t,moments='mvsk')
r = gamma.rvs (K,loc=0,scale=t, size=N)
# Grafico Experimentos
plt.plot(r)
plt.title(str(N)+" puntos distribuidos gamma con parametro= "+str(k))
plt.show()
r.sort()
# Grafico Experimentos ordenados
plt.plot(r)
plt.title(str(N)+" puntos distribuidos gamma con parametro ="+str(k))
plt.show()
# Histograma
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos  gamma con parametro= "+str(k))
plt.show()
# Datos estadisticos
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la gamma son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

La función gamma suele ser utilizada para modelar problemas que se relacionan con el tiempo, ya sean tareas que ocurran en un momento o tiempos de espera entre eventos. A partir de sus datos de escala y forma puede parecerse a una normal o a una exponencial. Para este caso tiene una gran similitud con una normal, con una cola mas larga de lado derecho.
